Machine Learning vs Deep Learning

<h2>Machine Learning</h2>
<li> the use and development of computer systems that are able to learn and adapt without following explicit instructions</li>
<li> Work on structured and label dataset</li>



<h1>Deep Learning</h1>
<li> Subset of Machine Learning</li>
<li> Does need a human to label the data set and no human intervention is needed</li>
<li> More complicated and layered structure of algorithm called Neural Networks
